```python
class Post:
    def __init__(self, title , body):
        self.title = title
        self.body  = body
    def __str__(self):
        return f"{self.title}"
    def __repr__(self):
        return self.__str__()
    
class Message:
    def __init__(self, from_User, body , to_User):
        self.body = body
        self.fromUser = from_User
        self.toUser = to_User
        self.status = "unviewed"
        
        
    def __str__(self):
        return f" Message:> {self.body}  From: {self.fromUser} -> To: <|> {self.toUser.nickname} <|>  Status: |-|-| {self.status} |-|-|"
        
    def __repr__(self):
        return self.__str__()
```
